```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv.imread("/home/willcmc/ml-lab/Day-2/nate.jpg", cv.IMREAD_GRAYSCALE)
```

```python
(height, width) = img.shape
```

```python
plt.imshow(img, cmap='gray');
```

## Fooling around with transposing

```python
plt.imshow(img.T, cmap='gray');
```

## Thresholding (MD vibes)

```python
thr = 100;
img_thresh = img > thr
plt.imshow(img_thresh, cmap='gray');
```

## Resizing to larger img

```python
img_resized = cv.resize(img, (2*width, 2*height), interpolation = cv.INTER_LINEAR)
plt.imshow(img_resized, cmap='gray');
```

## Affine Transform

```python
rows,cols = img.shape[:2]
tx = 500
ty = 500
M = np.float32([[1,0,tx],[0,1,ty]])
"""
M is transformation matrix.
"""
img_atf1 = cv.warpAffine(img, M, (cols,rows))
plt.imshow(img_atf1, cmap='gray');
```

## Flipping!

```python
img_refl = cv.flip(img, 0)
plt.imshow(img_refl, cmap='gray');
```

## Generating some motion (directional) blur in reflected image and concat it

```python
kernel_size = 20
kernel = np.zeros((kernel_size, kernel_size))
kernel[:, int((kernel_size - 1)/2)] = np.ones(kernel_size)
kernel /= kernel_size

img_refl_blur = cv.filter2D(img_refl, -1, kernel)

fade_mask = np.zeros((height,width))
img_total = np.concatenate((img, img_refl_blur), axis=0)
import matplotlib
matplotlib.rcParams['figure.figsize'] = [10, 10]
plt.imshow(img_total, cmap='gray');
```
