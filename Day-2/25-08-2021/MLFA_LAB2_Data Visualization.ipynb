# DATA VISUALIZATION

Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.

Reading such large datasets is impractical. Plots however give us a very good idea about the trend of the data and what approach should we follow while applying ML algorithms.

There are 2 very important libraries in python for data visualization :

*   matplotlib
*   seaborn

In matplotlib, we import pyplot which has a collection of popular functions in matplotlib. Its functions manipulate elements of a figure, such as creating a figure, creating a plotting area, plotting lines, adding plot labels, etc.

The seaborn package is used to create more attractive and informative statistical graphics.


```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

%matplotlib inline is required to make our plots appear and be stored within the notebook.

We will be working on a new dataset today which will have only 2 columns. One will be the input column while the other will be the output column. It is a linear regression dataset just to understand the basics of visualization.

Link to the dataset --> https://www.kaggle.com/andonians/random-linear-regression

```python
#from google.colab import files
#uploaded = files.upload()
import io
df = pd.read_csv('LR_Basics.csv',encoding = 'unicode_escape')
df.head()
```

Let us try to see how the output depends on the input by plotting it. We will be mostly using the matplotlib library to plot any type of graph but sometimes we will use the seaborn library. 

```python
plt.scatter(df['x'],df['y'])
plt.xlabel("x ----->")
plt.ylabel("y ----->")
plt.grid()
plt.legend(["x"])
plt.title("PLOT")
```

The **scatter** function plots all the points of the dataset on the graph. **xlabel** and **ylabel** allows us to label the x and y axis respectively. **title** provides a heading to the graph and **grid** function turns on the grid. The **legend** function allows us to provide guide to the representation in graph. It is useful if there are multiple plots in the same figure. If we write multiple plots in the same cell, they get plotted in the same figure. 

We can now understand that why do we prefer to plot the dataset. We can easily see the relationship between the output and the input. The output is linearly dependent on the input. We will learn how to find the most fitting line to the plot in the future lectures.

We can change various features like the color of the plot according to our requirement. 

```python
plt.scatter(df['x'],df['y'],c='r',marker='+')
plt.xlabel("x ----->")
plt.ylabel("y ----->")
plt.grid()
plt.title("PLOT")
```

Here we have changed the color of the plot to red and the display points to '+'. We can make many such changes in matplotlib library. You should check out the different types of functions and there parameters available in the library. Here is the link:

https://matplotlib.org/

Next, we will see how to plot a histogram using matplotlib. There are many such functions in matplotlib and here I have shown some.

```python
plt.hist(df['x'])
```

All the different plots have different significance and are useful in different situations. Here the histogram does not tell us much but it could be really helpful in other cases. Data visualization is one of the crucial topic in Machine Learning.
